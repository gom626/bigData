```python
import csv
import math
import matplotlib.pyplot as plt
```

```python
f=open("gender.csv",encoding="CP949")
data=csv.reader(f,delimiter=",")
```

```python
header=next(data)
#print(header)
```

```python
male=[]
female=[]
size=[]
```

```python
local=input("찾고 싶은 지역의 이름을 알려주세요 : ")
```

```python
for row in data:
    if local in row[0]:
        size.append(int(row[1]))
        for i in row[3:104]:
            male.append(int(i))
        size.append(int(row[104]))
        for i in row[106:]:
            female.append(int(i))
        break
```

```python
print(size)
```

```python
plt.barh(range(101),male)
plt.show()
```

```python
plt.rc('font',family="AppleGothic")
color=['crimson','darkcyan']
plt.axis('equal')
plt.pie(size,labels=['남','여'],autopct='%.1f%%',colors=color,startangle=90)
plt.title(local+' 지역의 남녀 성별 비율')
plt.show()
```

```python
diff=[]
ssize=[];
for i in range(101):
    diff.append(male[i]-female[i])
    ssize.append(math.sqrt((male[i]+female[i])))
```

```python
print(diff)
```

```python
plt.bar(range(101),diff)
plt.rcParams['axes.unicode_minus']=False
plt.show()
```

```python
plt.style.use('ggplot')
plt.rc('font',family="AppleGothic")
plt.figure(figsize=(10,5),dpi=300)
plt.scatter(female,male,s=ssize,alpha=0.5,c=range(101),cmap='jet')
plt.plot(range(max(male)),range(max(male)))
plt.colorbar()
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
