```python
import csv;
import matplotlib.pyplot as plt
```

```python
f=open('age.csv','r',encoding='cp949')
data=csv.reader(f,delimiter=',')
```

```python
header=next(data);
```

```python
string=input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ")
```

```python
result=[]
for row in data:
    if string in row[0]:
        for i in row[3:]:
            result.append(int(i));
print(result)
```

```python
plt.style.use('ggplot')
plt.rc('font', family='AppleGothic') 
plt.title(string+" 지역의 인구구조")
plt.plot(result)
plt.show()
```
