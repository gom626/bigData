```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
import csv
f=open('seoul.csv','r',encoding='cp949')
data=csv.reader(f,delimiter=',')
```

```python
header=next(data)
```

```python
date="";
temper=0;
Ltem=[]
Utem=[]
for row in data:
    temp=row[0][5:]
    year=int(row[0][:4])
    if(temp=="02-14" and year > 1983):
        if(row[-2]=="" or row[-1]==""):
            continue
        row[-2]=float(row[-2])
        row[-1]=float(row[-1])
        Ltem.append(row[-2])
        Utem.append(row[-1])
```

```python
plt.rc('font', family='AppleGothic') 
plt.rcParams['axes.unicode_minus']=False
plt.title('내 생일의 기온 변화 그래프')
plt.plot(Ltem,label="최저기온")
plt.plot(Utem,"hotpink",label="최고기온")
plt.legend()
plt.show()
```
